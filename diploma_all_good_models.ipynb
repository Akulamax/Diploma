```python
import pandas as pd
data = pd.read_csv(r'C:\Users\misha\Desktop\total_dataset_to_ML.csv')
data
```

```python
data1 = data
data2 = data[data['target'].isin([0, 1, 2, 4])]
print(len(data1))
print(len(data2))
```

```python
X1 = data1.drop(['target'], axis=1)
X2 = data2.drop(['target'], axis=1)
y1 = data1['target']
y2 = data2['target']
```

```python
from sklearn.model_selection import train_test_split

X1_train, X1_test, y1_train, y1_test = train_test_split(X1, y1, test_size=0.25,random_state=42)
X2_train, X2_test, y2_train, y2_test = train_test_split(X2, y2, test_size=0.25,random_state=42)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.model_selection import StratifiedShuffleSplit
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import cross_val_score
from sklearn.metrics import recall_score
from datetime import datetime
from imblearn.over_sampling import SMOTE
from imblearn.pipeline import Pipeline
import joblib

models = []
models.append(('SVM_all', SVC(C=0.1, gamma=0.1, kernel='poly')))
models.append(('SVM_all_smote', SVC(kernel='rbf')))
models.append(('SVM_half', SVC(C=0.1, gamma=0.1, kernel='poly')))
models.append(('SVM_half_smote', SVC(kernel='rbf')))

models.append(('KNN_all', KNeighborsClassifier(n_neighbors=1)))
models.append(('KNN_all_smote', KNeighborsClassifier(n_neighbors=12)))
models.append(('KNN_half', KNeighborsClassifier(n_neighbors=1)))
models.append(('KNN_half_smote', KNeighborsClassifier(n_neighbors=7)))


models.append(('DT_all', DecisionTreeClassifier(max_depth=18, max_features=0.3)))
models.append(('DT_all_smote', DecisionTreeClassifier(max_depth=8, max_features=0.2)))
models.append(('DT_half', DecisionTreeClassifier(max_depth=12, max_features=0.2)))
models.append(('DT_half_smote', DecisionTreeClassifier(max_depth=8, max_features=0.3)))


models.append(('RF_all', RandomForestClassifier(max_depth=10, max_features=0.2, n_estimators=300)))
models.append(('RF_all_smote', RandomForestClassifier(max_depth=8, max_features=0.2, n_estimators=100)))
models.append(('RF_half', RandomForestClassifier(max_depth=12, max_features=0.2, n_estimators=100)))
models.append(('RF_half_smote', RandomForestClassifier(max_depth=10, max_features=0.2, n_estimators=100)))


df_all=pd.DataFrame()
df_all['target'] = y1_test
              
df_half=pd.DataFrame()
df_half['target'] = y2_test
              
oversampler = SMOTE(sampling_strategy='not majority')

for name, model in models:
    tmp = name.split('_')
    if len(tmp) > 2:
        if tmp[1] == 'all':
            X_train, y_train = oversampler.fit_resample(X1_train, y1_train)
            X_test, y_test = X1_test, y1_test
        else:
            X_train, y_train = oversampler.fit_resample(X2_train, y2_train)
            X_test, y_test = X2_test, y2_test
    else:
        if tmp[1] == 'all':
            X_train, y_train = X1_train, y1_train
            X_test, y_test = X1_test, y1_test
        else:
            X_train, y_train = X2_train, y2_train
            X_test, y_test = X2_test, y2_test

    start_time = datetime.now()
    print(name, start_time)
    model.fit(X_train, y_train)
    filename = f'C:/Users/misha/Desktop/diploma_models/{name}.joblib'
    model = joblib.load(open(filename, 'rb'))
    pred = model.predict(X_test)
    print(confusion_matrix(y_test, pred))
    print(classification_report(y_test, pred))
    print(name, datetime.now() - start_time)
```
