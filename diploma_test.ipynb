```python
import pandas as pd
import numpy as np 
from scipy.stats import spearmanr
from scipy.stats import pearsonr
from scipy.stats import kendalltau
```

```python
data = pd.read_csv('/content/table_MEL_NV_1.csv')
data = data.drop(columns = ['Unnamed: 0'])
data.head(5)
```

```python
data.describe()
```

```python
std_zero_columns = []
for i, column in enumerate(data.columns):
  if data[column].std() == 0:
    std_zero_columns.append(column)
data_std = data.drop(columns=std_zero_columns)
data_std.head(5)
```

```python
df_spearmanr = pd.DataFrame(columns = ['rho', 'p'])
df_spearmanr
for i, column in enumerate(data_std.columns):
  df_spearmanr.loc[len(df_spearmanr)] = spearmanr(data[column], data['target'])
```

```python
df_spearmanr.head(3)
```

```python
df_spearmanr[df_spearmanr['rho'] > 0.35]
```

```python
df_spearmanr[df_spearmanr['rho'] < -0.4]
```

```python
df_pearsonr = pd.DataFrame(columns = ['r', 'p'])
df_pearsonr
for i, column in enumerate(data_std.columns):
  df_pearsonr.loc[len(df_pearsonr)] = pearsonr(data[column], data['target'])
```

```python
df_pearsonr.head(3)
```

```python
df_pearsonr[df_pearsonr['r'] > 0.35]
```

```python
df_pearsonr[df_pearsonr['r'] < -0.4]
```

```python
df_kendalltau = pd.DataFrame(columns = ['k', 'p'])
df_kendalltau
for i, column in enumerate(data_std.columns):
  df_kendalltau.loc[len(df_kendalltau)] = kendalltau(data[column], data['target'])
```

```python
df_kendalltau.head(3)
```

```python
df_kendalltau[df_kendalltau['k'] > 0.35]
```

```python
df_kendalltau[df_kendalltau['k'] < -0.35]
```
