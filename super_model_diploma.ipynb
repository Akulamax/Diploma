```python
def check_intersect_in_predict_models(data):
  all_rows = set()
  for row in data.drop(['target'], axis=1).columns:
    all_rows = all_rows.union(set(data[data['target'] == data[row]].index))
  return all_rows
```

```python
print(len(check_intersect_in_predict_models(data_half)), len(check_intersect_in_predict_models(data_all)))
```

```python
import numpy as np
def frequency_analysis(data):
  result_df = pd.DataFrame()
  for lesion_class in data['target'].unique():
    res = []
    for row in data[data['target'] == lesion_class].drop(['target'], axis=1).values:
      res.append((row == lesion_class).sum())
    result_df[f'{lesion_class}'] = pd.Series(res)
  return result_df
frequence_dataset_all = frequency_analysis(data_all)
frequence_dataset_half = frequency_analysis(data_half)
```

```python
def counts_prediction(dataset):
  df = pd.DataFrame(columns=['0', '1', '2', '3', '4', '5', '6', '7'])
  for index, i in enumerate(dataset.drop(['target'], axis=1).values):
    df.loc[index] = [(i == 0).sum(), (i == 1).sum(), (i == 2).sum(), (i == 3).sum(), (i == 4).sum(), (i == 5).sum(), (i == 6).sum(), (i == 7).sum()]
  df['target'] = dataset['target']
  return df
df_check = counts_prediction(data_all)
```

```python
for i in range(8):
  a = df_check[(df_check['target'] == i)]
  b = df_check[(df_check['target'] != i)]
  print(f'количество поражений {i} класса:', len(a))
  print(f'количество ложно нераспознанных поражений {i} класса: ', len(a[a[str(i)] == 0]))
  print(f'количество верно нераспознанных поражений {i} класса: ', len(b[b[str(i)] == 0]))
  res_tp = []
  res_fp = []
  for j in range(1, 9):
    count_tp = len(a[a[str(i)] == j])
    count_fp = len(b[b[str(i)] == j])
    print(f'количество верно распознанных поражений {j} моделями: ', count_tp)
    print(f'количество ложно распознанных поражений {j} моделями: ', count_fp)
    res_tp.append(count_tp)
    res_fp.append(count_fp)
  for k in range(0, len(res_tp)):
    sum_tp = 0
    sum_fp = 0
    for l in range(k, 8):
      sum_tp += res_tp[l]
      sum_fp += res_fp[l]
    if sum_tp + sum_fp == 0:
      print('нет распознанных поражений')
    else:
      print(f'количество верно распознанных поражений: {sum_tp}, количество ложно распознанных {sum_fp}, для {k+1} и большего числа моделей')
      print(f'отношение количества верно распознанных поражений к распознанным пораежниям, для {k + 1} и большего числа моделей: ', f'{(sum_tp / (sum_tp + sum_fp)):.2f}')
  print('\n')
```

```python
def relation_zero_to_top(column, top):
  return (column == 0).sum() / (column > top).sum()

def take_all_relation_with_value(data, value):
  for column in data.columns:
    top = 0
    if top != 7 and relation_zero_to_top(data[column], top) < value:
      top += 1
    print(column, top, relation_zero_to_top(data[column], top))

take_all_relation_with_value(frequence_dataset_half, 0.3)

#frequence_dataset_all['1'].value_counts()
```

```python
def take_counts_df(dataset):
  df = pd.DataFrame()
  for column in dataset.columns:
     df[column] = dataset[column].value_counts(ascending=True)
  return df

counts_df_all = take_counts_df(frequence_dataset_all)
counts_df_half = take_counts_df(frequence_dataset_half)
```
