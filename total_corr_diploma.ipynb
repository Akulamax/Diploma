```python
import pandas as pd
data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/total_dataset_to_corr.csv')
data
```

```python
data = data.drop(['Unnamed: 0', 'image_name'], axis = 1)
def change_name_lesion(name):
  if name == 'MEL':
    return 0
  elif name == 'NV':
    return 1
  elif name == 'BCC':
    return 2
  elif name == 'AK':
    return 3
  elif name == 'BKL':
    return 4
  elif name == 'DF':
    return 5
  elif name == 'SCC':
    return 6
  elif name == 'VASC':
    return 7

data['target'] = data['target'].apply(change_name_lesion)
data.head(3)
```

```python
data_MEL = data[data['target'] == 0] 
data_NV = data[data['target'] == 1]
data_BCC = data[data['target'] == 2]
data_AK = data[data['target'] == 3]
data_BKL = data[data['target'] == 4]
data_DF = data[data['target'] == 5]
data_SCC = data[data['target'] == 6]
data_VASC = data[data['target'] == 7]
```

```python
data_VASC
```

```python
from datetime import datetime

def get_corr_dataframe(dataframe):
    df_spearmanr = pd.DataFrame(columns = ['rho', 'p'])
    target = dataframe['target']
    dataframe = dataframe.drop(['target'], axis=1)
    for i, column in enumerate(dataframe.columns):
      df_spearmanr.loc[len(df_spearmanr)] = spearmanr(dataframe[column], target)
    return df_spearmanr[np.abs(df_spearmanr['rho']) > np.quantile(df_spearmanr['rho'], 0.999)]

def clean_dataframe(dataframe):
  std_zero_columns = []
  for i, column in enumerate(dataframe.columns):
    if dataframe[column].std() == 0:
      std_zero_columns.append(column)
  print('классы: ', dataframe['target'].unique()[0], dataframe['target'].unique()[1])
  print('количеств столбцов с нулевым std: ', len(std_zero_columns))
  data_std = dataframe.drop(columns=std_zero_columns)
  return data_std

list_datasets = [data_MEL, data_NV, data_BCC, data_AK, data_BKL, data_DF, data_SCC, data_VASC]
list_datasets_of_corr = []
set_of_column_names = set()

start = datetime.now()
for i in range(len(list_datasets)):
  for j in range(i + 1, len(list_datasets)):
      dataframe = pd.concat([list_datasets[i], list_datasets[j]])
      data_std = clean_dataframe(dataframe)
      df_corr = get_corr_dataframe(data_std)
      for k in df_corr.index:
        set_of_column_names.add(data_std.iloc[:,[k]].columns[0])
      print('количество строк в датафрейме коэффициентов: ', len(df_corr))
      list_datasets_of_corr.append(df_corr)
print('время работы: ', datetime.now() - start)
```

```python
for index, i in enumerate(list_datasets_of_corr):
  maximum = np.abs(i.drop(i.index[-1]).rho).max()
  minimum = np.abs(i.drop(i.index[-1]).rho).min()
  print(f'номер датасета: {index}; max: {maximum:.3f}; min: {minimum:.3f}; разность: {(maximum - minimum):.3f}')
```
